# Project Title: Multi-Dimensional Competition Risk Model for Aramco’s Overseas Markets

**Project Description**

This project analyzes the rising global competition faced by Aramco in the international energy market. Using a 10,000-record, 40-feature dataset, the study evaluates market share, pricing dynamics, LNG pressure, renewable energy impact, logistics disadvantages, and regional competition. The objective is to identify how national oil companies, traders, and renewable suppliers influence Aramco’s global operations. Insights from multivariate analysis help highlight risk levels, competitive gaps, and strategic actions to strengthen Aramco’s global market position.

**Problem Statement**
 - Aramco’s Global & International Operations face increasing pressure due to intensifying competition in the global energy market. National oil companies, LNG exporters, multinational traders, and rapidly expanding renewable energy producers are capturing market share and influencing pricing, demand patterns, and customer preferences. This competitive landscape creates challenges in maintaining sales volumes, optimizing supply routes, and sustaining long-term contracts across regions such as Asia, Europe, India, and North America.
 - To address this issue, a detailed analytical framework is required to understand competitive gaps, evaluate risk levels, and identify strategic opportunities that can strengthen Aramco’s global market position.
